```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
plt.show()
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
```

```python
plt.show()
```

```python
plt.subplot(1,2,1) #altgrafik 1-> kaç sıra olacağı 2-> kaç grafik gösterileceği 1->1.'nin çizeleceği
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) #1 sıra olacak, 2 kolon olacak, şu an 2.grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
plt.show(
)
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2, 0.9,0.9]) #0.2 -> sol taraftaki eksenin başlangıç noktası , 0.2 -> sağ taraftaki eksenin başlangıç noktası
#0.9 -> genişliği, 0.9 -> uzunluğu
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
plt.show()
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.1,0.1, 0.6,0.6]) #0.2 -> sol taraftaki eksenin başlangıç noktası , 0.2 -> sağ taraftaki eksenin başlangıç noktası
#0.9 -> genişliği, 0.9 -> uzunluğu
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
plt.show()
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"g")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
plt.show()
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"g")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
plt.show()
```

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"g")
plt.show()
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1,ncols=2) #1 satır 2 sütun olacak şekilde
#benimFigure: Bu, genel figür (grafik alanı) nesnesidir.
#benimEksenler: Bu, alt grafik eksenlerinin bir listesi (ya da dizisi) olacak. Yani her bir eksende bir grafik olacak.

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"r")
    eksen.set_xlabel("X Ekseni")
plt.show()
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"r")
    eksen.set_xlabel("X Ekseni")
plt.tight_layout()
plt.show()
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=2,ncols=2)
type(benimEksenler)

for eksen in benimEksenler.flat: #.flat özelliği, bu diziyi tek boyutlu hale getirir, böylece her bir eksene tek tek erişebilirsin.
    eksen.plot(numpyDizisi1,numpyDizisi2,"r")
    eksen.set_xlabel("X Ekseni")
plt.tight_layout()
plt.show()
```

```python
yeniFigur = plt.figure()
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend() #legend(), bu isimleri içeren bir kutu ekleyerek hangi çizginin hangi veriyi temsil ettiğini anlamamızı sağlıyor.
```
